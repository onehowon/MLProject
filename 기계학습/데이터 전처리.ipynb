### 구성원수 컬럼의 속성값이 6개이므로 세분화(NaN은 0으로 표현)

```python
def categorize_member_count(count_str):
    if count_str == '11-50명':
        return 1
    elif count_str == '51-200명':
        return 2
    elif count_str == '201-500명':
        return 3
    elif count_str == '501-1000명':
        return 4
    elif count_str == '1001-5000명':
        return 5
    else:
        return 0  # NaN 값 처리

# '구성원수' 컬럼을 이용하여 새로운 컬럼 생성
df['구성원수_분류'] = df['구성원수'].apply(categorize_member_count)
```

```python
df.drop(columns=['구성원수'], inplace=True)
df.drop(columns=['구성원수_분류'], inplace=True)
```

### 기술스택 개수 수치화

```python
# 기술 스택 열을 문자열로 변환하고 "|" 기호를 기준으로 분할하여 리스트로 변환
tech_stack_list = df['기술스택명'].astype(str).apply(lambda x: x.split("||"))

# 각 행의 기술 스택 리스트의 길이를 계산하여 새로운 열에 저장
df['기술스택_개수'] = tech_stack_list.apply(len)

# 기술 스택 개수의 평균 계산
average_tech_stack_count = df['기술스택_개수'].mean()

# 평균보다 크거나 같은 기술 스택 개수를 가진 행 선택
filtered_df = df[df['기술스택_개수'] >= average_tech_stack_count]
```

```python
df.drop(columns=['기술스택명'], inplace=True)
```

### 산업구분 구체화

```python
df['산업구분명'] = df['산업구분명'].fillna('')
```

```python
def categorize_industry_by_keyword(industry):
    if '소프트웨어' in industry or '데이터' in industry or '인공지능' in industry or 'Clud' in industry or '서버' in industry or '웹' in industry or '모바일' in industry or 'AI' in industry or 'AR' in industry or 'Unity' in industry or 'c#' in industry or '솔루션' in industry or 'mobile' in industry or 'platform' in industry or 'IT' in industry:
        return 'IT'
    else:
        return 'None IT'

# 기존의 산업구분명을 새로운 범주로 변환하여 새로운 열에 저장
df['산업구분명_분류'] = df['산업구분명'].apply(categorize_industry_by_keyword)
```

```python
df.drop(columns=['산업구분명'], inplace=True)
```

## 본사 위치 이진화(수도권, 비수도권)

```python
def categorize_industry_by_region(region):
    if '서울특별시' in region or '경기도' in region:
        return '수도권'
    else:
        return '비수도권'

df['지역명'] = df['지역명'].apply(categorize_industry_by_region)
```
