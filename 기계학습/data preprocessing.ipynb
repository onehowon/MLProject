```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
company_information = pd.read_csv("csv_tb_rk_entrp_info_m.csv", encoding='cp949')
```

```python
columns_to_drop = ["업체국문명","업체영문명", "기업소개상세내용", "홈페이지url", "소셜미디어url", "기업추천내용", "사무실주소", "발행사명", "보도일자", "기사url"]
```

```python
company_information_revised = company_information.drop(columns=columns_to_drop) 
```

```python
service_product_information = pd.read_csv("csv_tb_rk_srvc_prduc_m.csv", encoding='euc-kr')
```

```python
columns_to_drop = ["제품일련번호", "제품영문명", "제품상세설명", "홈페이지url"]
```

```python
service_product_information_revised = service_product_information.drop(columns = columns_to_drop)
```

```python
region_investment_information = pd.read_csv("csv_tb_rk_lct_invsm_info_m.csv", encoding='euc-kr')
```

```python
columns_to_drop = ["지역별투자정보일련번호", "최종투자유치일자", "최종투자유치단계명", "누적투자금액"]
```

```python
region_investment_information_revised = region_investment_information.drop(columns=columns_to_drop)
```

```python
total_investment_information = pd.read_csv("csv_tb_rk_invsm_attrt_m.csv", encoding="euc-kr")
```

```python
columns_to_drop = ["투자일련번호", "보도자료발행사명", "보도자료제목", "보도자료url", "투자기관번호"]
```

```python
total_investment_information_revised = total_investment_information.drop(columns = columns_to_drop)
```

```python
company_information_revised.head(10)
```

```python
# 각 데이터프레임의 '기업일련번호' 컬럼에 중복된 값이 있는지 확인
print("service_product_information_revised 중복 확인:", service_product_information_revised['기업일련번호'].duplicated().any())
print("region_investment_information_revised 중복 확인:", region_investment_information_revised['기업일련번호'].duplicated().any())
print("total_investment_information_revised 중복 확인:", total_investment_information_revised['기업일련번호'].duplicated().any())
print("company_information_revised 중복 확인:", company_information_revised['기업일련번호'].duplicated().any())
```

```python
service_product_information_revised.drop_duplicates(subset=['기업일련번호'], keep='first', inplace=True)
region_investment_information_revised.drop_duplicates(subset=['기업일련번호'], keep='first', inplace=True)
total_investment_information_revised.drop_duplicates(subset=['기업일련번호'], keep='first', inplace=True)
```

```python
# 각 데이터프레임의 '기업일련번호' 컬럼에 중복된 값이 있는지 확인
print("service_product_information_revised 중복 확인:", service_product_information_revised['기업일련번호'].duplicated().any())
print("region_investment_information_revised 중복 확인:", region_investment_information_revised['기업일련번호'].duplicated().any())
print("total_investment_information_revised 중복 확인:", total_investment_information_revised['기업일련번호'].duplicated().any())
print("company_information_revised 중복 확인:", company_information_revised['기업일련번호'].duplicated().any())
```

```python
# 첫 번째 병합: company_information_revised와 service_product_information_revised 병합
total_information = pd.merge(company_information_revised, service_product_information_revised, on="기업일련번호", how="inner")

# 두 번째 병합: total_information과 region_investment_information_revised 병합
total_information = pd.merge(total_information, region_investment_information_revised, on="기업일련번호", how="inner")

# 세 번째 병합: total_information과 total_investment_information_revised 병합
total_information = pd.merge(total_information, total_investment_information_revised, on="기업일련번호", how="inner")
```
